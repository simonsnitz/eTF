```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# define plotting function
sns.set()
sns.set_style("white")
plt.rcParams['font.sans-serif'] = ['Arial'] 
sns.set_style("ticks", {'xtick.direction':'in', 'xtick.top':False, 'ytick.direction':'in', 'ytick.right':False })
plt.rcParams["xtick.labelsize"] = 20
plt.rcParams["ytick.labelsize"] = 0

def plot_chromatogram(x, y, color, name):
    fig = plt.figure(figsize=(10,4))
    ax = fig.add_subplot()
    ax.spines[['right', 'top']].set_visible(False)
    plt.ticklabel_format(style="plain", axis="x")
    plt.xticks(np.arange(0, 11, step=1))
    
    plt.xlabel("Time (min)", fontsize=36, labelpad=15)
    # plt.ylabel("Count", fontsize=24, labelpad=15)

    plt.plot(x, y, color=color, linewidth=3)
    # plt.savefig(name, format="svg")
    plt.show()
```

```python
# load data
chrom = "IR35-1"
with open("../Data/IR35_Chromatograms.xlsx", 'rb') as f:
    data = pd.read_excel(f, sheet_name=chrom)

x = data["Time (min)"][200:700]
y = data["Count"][200:700]

color = "green"
name = "SVG/IR35-1_Chromatogram.svg"

plot_chromatogram(x,y,color, name)
```
