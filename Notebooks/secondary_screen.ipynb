```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statistics import mean
```

```python
# define plotting function

def plot_secondary_screen(x, y, WT, title):
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_subplot()
    plt.xticks(rotation=90, size=10)
    plt.yticks(size=14)
    #calculate max value
    z = list(y)
    z.append(WT)
    max_value = max(z)
    #set ylim to be 20% above max value
    plt.ylim(0,max_value*1.2)

    plt.title(title, fontsize=18)
    plt.xlabel("IRED variant", fontsize=18, labelpad=15)
    plt.ylabel("Fold change with DHIQ", fontsize=18, labelpad=15)
    plt.bar(x, y, edgecolor="#000000", color="#85afff", width=0.7)
    plt.axhline(y=WT, color='red', linestyle='--', linewidth=2)

    # plt.savefig("SVG/"+str(title)+".svg", format="svg")
    plt.show()
```

```python
# load IRED-R1 pScreenS3 data
with open("../Data/Secondary_screen.xlsx", 'rb') as f:
    data = pd.read_excel(f, sheet_name="R1_pScreenS")

# Get variant labels and fold-change values. Skip WT
x = data["Name"][0:-4]
y = data["Fold_change"][0:-4]
title = "R1 pScreenS3"

# calculate average WT
WT = mean(data["Fold_change"][-4:])

plot_secondary_screen(x,y, WT, title)
```
